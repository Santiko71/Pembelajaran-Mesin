```python
import pandas as pd
import bz2
from langdetect import detect
from tqdm import tqdm
import ftfy
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

# Ubah path file sesuai dengan lokasi file Anda
file_path = r"D:/_KULIAH/#SEMESTER4/Pembelajaran Mesin (Machine Learning)/UTS/archive/test.ft.txt.bz2"

# Membaca file .bz2 dengan memory_map
with bz2.open(file_path, "rt", encoding="utf-8") as f:  # Menentukan encoding utf-8
    lines = f.readlines()

# Menginisialisasi list untuk menyimpan hasil
results = []

# Loop melalui setiap baris dengan tqdm
for line in tqdm(lines):
    # Pisahkan label kelas dan teks ulasan
    label, text = line.split(maxsplit=1)

    # Bersihkan teks non-Latin
    text = ftfy.fix_text(text)

    # Deteksi bahasa pada teks ulasan
    try:
        language = detect(text)
    except:
        language = 'N/A'

    # Tambahkan hasil ke dalam list
    results.append((label, text, language))

# Membuat DataFrame dari hasil
data = pd.DataFrame(results, columns=["Label", "Review", "Language"])

# Klasifikasi ulasan menjadi rating tinggi (4 dan 5) dan rating rendah (1 dan 2)
data['Sentiment'] = data['Label'].apply(lambda x: 'Positive' if x in ['__label__4', '__label__5'] else 'Negative')

# Membagi data menjadi data latih dan data uji
train_texts, test_texts, train_labels, test_labels = train_test_split(data['Review'], data['Sentiment'], test_size=0.2, random_state=42)

# Menghitung distribusi kelas dalam data latih
class_distribution = train_labels.value_counts(normalize=True)

# Membuat dan melatih model Naive Bayes dengan menggunakan distribusi kelas yang dinormalisasi
model = MultinomialNB(class_prior=class_distribution.values)
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(train_texts)
X_test = vectorizer.transform(test_texts)
model.fit(X_train, train_labels)

# Memprediksi label pada data uji
predictions = model.predict(X_test)

# Menampilkan laporan klasifikasi
print(classification_report(test_labels, predictions))
```
